```python
plt.figure(figsize=(7, 7))
df['sentiment'].value_counts().plot.pie(autopct='%1.1f%%', startangle=140, colors=sns.color_palette("tab10"))
plt.title('Distribution of Sentiment Classes')
plt.ylabel('')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Duygu sÄ±nÄ±flarÄ±nÄ±n sayÄ±sÄ±nÄ± hesapla
sentiment_counts = df['sentiment'].value_counts()

plt.figure(figsize=(10, 6))
sns.barplot(x=sentiment_counts.index, y=sentiment_counts.values, palette="viridis")

# Grafik baÅlÄ±klarÄ±nÄ± ekle
plt.title('Distribution of Sentiment Classes')
plt.xlabel('Sentiment')
plt.ylabel('Number of Tweets')

# Her Ã§ubuÄa sayÄ± etiketleri ekle
for index, value in enumerate(sentiment_counts.values):
    plt.text(index, value, str(value), ha='center', va='bottom', fontsize=12, fontweight='bold', color='black')

# X eksenindeki etiketlerin daha kolay okunabilmesi iÃ§in dÃ¶ndÃ¼r
plt.xticks(rotation=45, ha='right')

# Izgara Ã§izgilerini ekle
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Grafik alanÄ±nÄ± sÄ±kÄ±laÅtÄ±r
plt.tight_layout()

# Grafik gÃ¶ster
plt.show()
```
