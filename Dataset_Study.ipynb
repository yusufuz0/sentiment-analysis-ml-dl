```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = '/content/drive/My Drive/Colab Notebooks/YapayZekaDonemSonuUygulaması(Emotion detection from text)/tweet_emotions.csv'
df = pd.read_csv(file_path)
```

```python
plt.figure(figsize=(7, 7))
df['sentiment'].value_counts().plot.pie(autopct='%1.1f%%', startangle=140, colors=sns.color_palette("tab10"))
plt.title('Distribution of Sentiment Classes')
plt.ylabel('')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Duygu sınıflarının sayısını hesapla
sentiment_counts = df['sentiment'].value_counts()

plt.figure(figsize=(10, 6))
sns.barplot(x=sentiment_counts.index, y=sentiment_counts.values, palette="viridis")

# Grafik başlıklarını ekle
plt.title('Distribution of Sentiment Classes')
plt.xlabel('Sentiment')
plt.ylabel('Number of Tweets')

# Her çubuğa sayı etiketleri ekle
for index, value in enumerate(sentiment_counts.values):
    plt.text(index, value, str(value), ha='center', va='bottom', fontsize=12, fontweight='bold', color='black')

# X eksenindeki etiketlerin daha kolay okunabilmesi için döndür
plt.xticks(rotation=45, ha='right')

# Izgara çizgilerini ekle
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Grafik alanını sıkılaştır
plt.tight_layout()

# Grafik göster
plt.show()
```
